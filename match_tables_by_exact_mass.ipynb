```r
# Load the tables and make sure the 
# table 'tab' is the reference table
# and has the mass in the second column
tab <- read.csv("Boiga_MS1.csv")
tab2 <- read.csv("Boiga_toppic.csv")
```

```r
# match the second column of the tab, to all masses in a column named "Precursor.mass"
# of tab2 in a window of 20 ppm (change the value below if needed)
lmatch20 <- lapply(tab[,2], function(x) which( (abs(tab2[,"Precursor.mass"]-x)/x)*10^6 < 20))
```

```r
# count hits
sum(unlist(lapply(lmatch20, length)))
```

```r
colnames(tab2)
```

```r
tab2[lmatch20[[2]],'Protein.name'][order(tab2[lmatch20[[2]],'P.value'])][1:10]
```

```r
head(tab2)
```

```r
# collapse the 'ID's in the tab2 by ';' and as a column to tab
match <- (lapply(lmatch20, function(x) if(length(x)) as.character(tab2[x,"ID"][order(tab2[x,'P.value'])]) else ""))
pname <- (lapply(lmatch20, function(x) if(length(x)) as.character(tab2[x,"Protein.name"][order(tab2[x,'P.value'])]) else ""))
pval <- (lapply(lmatch20, function(x) if(length(x)) tab2[x,"P.value"][order(tab2[x,'P.value'])] else ""))
proteoform <- (lapply(lmatch20, function(x) if(length(x)) tab2[x,"Proteoform"][order(tab2[x,'P.value'])] else ""))    

```

```r
length(match)
length(pname)
length(pval)
length(proteoform)
```

```r
# select the n first values, sorted by pval
match <- unlist(lapply(match, function(x) if(length(x) > 1) paste(x[1], collapse = ";") else paste(x, collapse = ";")))
pname <- unlist(lapply(pname, function(x) if(length(x) > 1) paste(x[1], collapse = ";") else paste(x, collapse = ";")))
pval <- unlist(lapply(pval, function(x) if(length(x) > 1) paste(x[1], collapse = ";") else paste(x, collapse = ";")))
proteoform <- unlist(lapply(proteoform, function(x) if(length(x) > 1) paste(x[1], collapse = ";") else paste(x, collapse = ";")))    
```

```r
length(match)
length(pname)
```

```r
tab <- cbind(tab, match, pname, pval, proteoform)
```

```r
head(tab)
```

```r
# write to a new file
write.csv(tab,  "Boiga_MS1_with_match_and_name_and_pval_and_proteoform.csv", row.names=FALSE)
```
